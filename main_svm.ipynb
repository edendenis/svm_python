#### CARREGAMENTO, ARMAZENAMENTO E MANIPULAÇÃO DE DADOS:

##### O banco de dados deve estar estruturado da seguinte forma:
    1. A primeira coluna deve ser a de nomeação/identificação das lições;
    2. As colunas seguintes, exceto a última, devem ser as variáveis do problema;
    3. A última coluna deve ser classificada com valores binários, por exemplo, - 1 (menos um) ou + 1 (mais um), de preferência valores numéricamente "distantes" (- 100 e + 100, por exemplo).

```python
# Carregamento:

# Arquivo Original:
# endereco: endereco do banco de dados
endereco = "banco_de_dados/Banco_de_Dados_Desjuntado_em_G_e_H.xlsm" 

# banco_de_dados: variável que armazena o banco de dados
banco_de_dados = pd.read_excel(endereco, 
                               sheet_name=0)

banco_de_dados.head()
```

```python
type(banco_de_dados)
```

```python
# Manipulação dos Dados:

"""
bd_auxiliar: banco de dados auxiliar, cópia do primeiro armazenamento, para
NÃO alterar o primeiro, caso seja necessário utilizá-lo posteriormente 
no código.
"""
bd_auxiliar = banco_de_dados.copy()
```

#### VARIÁVEL(IS):

```python
# valor_c: coeficiente de ajuste 
valor_c = 1
# input("Inserir o valor desejado para o coeficiente " + 
# "de ajuste 'c':")
```

```python
print("O valor do 'Coeficiente de Ajuste' é = " + str(valor_c))
```

#### CÁLCULO(S):

```python
# Remover a primeira coluna:
bd_auxiliar = bd_auxiliar.drop([bd_auxiliar.columns[0]], 
                               axis=1)
bd_auxiliar.head()
```

```python
"""
Os objetos estão sendo salvos para NÃO ter que
executar todo o código novamente, simplesmente
para executar uma function ou fazer um servidor
ler o arquivo.
"""

# Escrever o arquivo pickle:
# :param protocol=4: possibilitar carregar objetos com mais de 4 GB. 
pkl.dump([bd_auxiliar,
         valor_c],
        open("pickles/objetos_svm.pkl", "wb"),
        protocol=4)
```

```python
[B, valores, vector_solucao, soma_vector_C] = svm.svm(bd_auxiliar, 
                                                      valor_c,
                                                      tolerancia=0.1)
```

```python
# Verificação apenas dos valores do banco de dados que foi gerada a soloção
pd.DataFrame(valores)
```

```python
print("Coeficientes de ajuste")
pd.DataFrame(vector_solucao) 
```

```python
# soma_vector_C = Coeficiente linear da Solução do modelo
print("Coeficiente Linear", soma_vector_C)
```

#### Resultado(s):

```python
y_teo = np.zeros(len(B))
for i in range(0, valores.shape[0], 1):
    y_teo[i] = np.matmul(valores[i], vector_solucao) + soma_vector_C
y_teo = np.where(y_teo > 0, 100, -100)
"""
A linha acima foi inserida para alterar as saídas positivas para +100 e as 
negativas para -100. Esta ação foi tomada para que o modelo possa entregar as 
respostas com apenas duas características. Os valores +100 e -100 foram 
escolhidos para que as duas saídas fiquem "distantes" o suficiente para o 
modelo poder decidir com mais facilidade as lições que sejam mais indefinidas.
"""
pd.DataFrame(y_teo)
```

```python
# Ler o arquivo gerado:
pkl.dump([B,
         y_teo],
        open("pickles/objetos_r_quadrado.pkl", "wb"),
        protocol=4)
```

```python
# R_quadrado: coeficiente de Acerto/Determinação
R_quadrado = r_quadrado.r_quadrado(B, y_teo)
print("R² =", R_quadrado * 100, "%")
```
